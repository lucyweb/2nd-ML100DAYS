試著使用 sklearn datasets 的其他資料集 (boston, ...)，來訓練自己的線性迴歸模型，並加上適當的正則話來觀察訓練情形。

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score  #指標 mean_squared_error
```

```python
#讀取波士頓房價資料集#
boston = datasets.load_boston()
```

```python
print(boston.DESCR)
```

```python
#讀取資料集
X = boston.data
len(X)
```

```python
#讀取目標
Y = boston.target
len(Y)
```

```python
# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2,random_state =4)
```

```python
#建立線性回歸
regr = linear_model.LinearRegression()

#將訓練資料(train)丟進去模型訓練; fit
regr.fit(x_train, y_train)

#predict y
y_pred = regr.predict(x_test)
```

```python
#取得回歸係數
print(regr.coef_)
```

```python
print("Mean_squared_error: %.2d" %mean_squared_error(y_test, y_pred))
```

# **LASSO**

```python
#讀入波士頓資料集
bos = datasets.load_boston()
```

```python
# 切分訓練集/測試集
x1_train, x1_test, y1_train, y1_test = train_test_split(bos.data, bos.target, test_size = 0.2, random_state = 4)
```

```python
# 建立一個線性回歸模型
lasso = linear_model.Lasso(alpha = 1.0)
```

```python
# 將訓練資料丟進去模型訓練
lasso.fit(x1_train, y1_train)

#predict y
y1_pred = lasso.predict(x1_test)
```

```python
# 印出各特徵對應的係數，可以看到許多係數都變成 0，Lasso Regression 的確可以做特徵選取
lasso.coef_
```

```python
# 預測值與實際值的差距，使用 MSE
print('MSE: %.2f' %mean_squared_error(y1_test, y1_pred))
```

 LASS 的結果並沒有比原本的線性回歸來得好， 這是因為目標函數被加上了正規化函數，讓模型不能過於複雜，相當於限制模型擬和資料的能力
 原來 MSE: 25 ; Lasso MSE: 28.95

# **Ridge**

```python
bo = datasets.load_boston()
```

```python
# 切分訓練集/測試集
x2_train, x2_test, y2_train, y2_test = train_test_split(bo.data, bo.target, test_size= 0.2, random_state= 4)
```

```python
# 建立一個線性回歸模型
ridge = linear_model.Ridge(alpha = 1.0)
```

```python
# 將訓練資料丟進去模型訓練
ridge.fit(x2_train, y2_train)

#predict y
y2_pred = ridge.predict(x2_test)
```

```python
# 印出各特徵對應的係數,可以很明顯看到比起 Linear Regression，參數的數值都明顯小了許多

ridge.coef_
```

```python
regr.coef_
```

```python
print("MSE: %.2d" %mean_squared_error(y2_test, y2_pred))
```

```python
print("regr MSE: %.2d" %mean_squared_error(y_test, y_pred))
print("lasso MSE: %.2d" %mean_squared_error(y1_test, y1_pred))
```

可以看見 LASSO 與 Ridge 的結果並沒有比原本的線性回歸來得好
