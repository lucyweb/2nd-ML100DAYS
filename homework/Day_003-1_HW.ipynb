### Hints: [隨機產生數值](https://blog.csdn.net/christianashannon/article/details/78867204)

```python
import pandas as pd
import numpy as np
```

```python
country = ['Taiwan','Japan','China','Veitnam','Tailand'] #國家
pop = np.random.randint(900000, size=5) #人口
```

```python
data = {'country': country,
       'population': pop}
```

```python
df = pd.DataFrame(data) 
df
```

```python
df['population'].max() #人口取最大值, 但沒有國家
```

```python
df[['country','population']].nlargest(1, 'population') #取到最大人口數國家
```
